```python
import numpy as np
import cv2
import matplotlib.image as matimg
from PIL import Image
import matplotlib.pyplot as plt
```

```python
#openCV - BGR - numpy array
img_cv = cv2.imread('./data/flying_bird.jpg')
#Matplotlib - RGB - numpy array
img_mat = matimg.imread('./data/flying_bird.jpg')
#PIL - RGB - object
img_pil = Image.open('./data/flying_bird.jpg')
```

```python
#openCV
print(img_cv)
```

```python
#matplotlib
print(img_mat)
```

```python
img_pil
```

```python
img_pil_arr = np.array(img_pil)
```

```python
img_pil_arr
```

```python
#display image using opencv
cv2.imshow('open_cv BGR',img_cv)
cv2.imshow('Matplotlib RGB',img_mat)
cv2.imshow('Pillow RGB',img_pil_arr)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#display image using matplotlib
plt.subplot(1,3,1)
plt.imshow(img_cv)
plt.axis('off')
plt.title('Matplotlib RGB')

plt.subplot(1,3,2)
plt.imshow(img_mat)
plt.axis('off')
plt.title('OpenCV BGR')

plt.subplot(1,3,3)
plt.imshow(img_pil_arr)
plt.axis('off')
plt.title('Pillow RGB')
plt.show()
```
